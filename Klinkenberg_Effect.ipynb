```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def klinkenberg():
  k_l=float(input('Enter the initial guess of absolute permeability in mD : '))
  p_m=float(input('Enter the mean pressure in psi : '))
  k_g=float(input('Enter the gas permeability in mD : '))

  while abs(6.9*(k_l**0.64)+p_m*k_l-p_m*k_g) >0.0001:
        k_l = k_l - (6.9*(k_l**0.64)+p_m*k_l - p_m*k_g)/(4.416*(k_l**(-0.36))+p_m)

  print('The value of absolute permeability after iteration or gas permeability at infinite mean pressure is : ',k_l)

  x=[0,1/p_m]
  y=[k_l,k_g]
  plt.figure(figsize=(8,4))
  plt.plot(x,y,color='red')
  plt.style.use('classic')
  plt.xlabel('1/Mean_pressure (psi^-1)')
  plt.ylabel('Permeability')
  plt.title('Klinkenberg Effect')
  plt.grid(True)

klinkenberg()
```
