```python
  import numpy as np
  import pandas as pd
  import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,100,size=20)
y=np.random.randint(1,90,size=20)
print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.show()
```

Elbow method-to find ideal value of k

```python
from sklearn.cluster import KMeans

data=list(zip(x,y))
intertias=[]

for i in range(1,21):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(data)
  intertias.append(kmeans.inertia_)

plt.plot(range(1,21),intertias,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Intertia/Cemtroids')
plt.show()
```

Model construction

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(data)

plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```
