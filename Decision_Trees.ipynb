Import libraries and dataset

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name="species")

df = pd.concat([X, y], axis=1)
df.head()
```

Split in train-test

```python
from sklearn.model_selection import train_test_split

x=df.drop('species',axis=1)
y=df['species']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
x_train2,x_validation,y_train2,y_validation=train_test_split(x_train,y_train,test_size=0.20)
```

Import decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

Apply DT using entropy, check accuracy

```python
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=3)
dtree.fit(x_train2,y_train2)

predictions = dtree.predict(x_validation)

from sklearn.metrics import accuracy_score
accuracy_score(y_validation,predictions)
```

Apply DT using gini, check accuracy

```python
dtree2=DecisionTreeClassifier(criterion='gini',max_depth=3)
dtree2.fit(x_train2,y_train2)

predictions2 = dtree2.predict(x_validation)

from sklearn.metrics import accuracy_score
accuracy_score(y_validation,predictions2)
```

Plotting

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig=plt.figure(figsize=(12,10))
a=plot_tree(dtree,feature_names=df.columns,fontsize=12,filled=True,class_names=['Setosa','Virginica','Versicolor'])
```
